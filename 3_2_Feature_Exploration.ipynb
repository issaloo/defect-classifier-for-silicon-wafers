```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

# 0. Import Data

```python
model_df = pd.read_csv('model_dataset/labeled_processed_data.csv')
```

```python
plt.style.use('fivethirtyeight')
```

# 1. Analyze relationships through graphs

```python
sns.countplot(x='class', data=model_df)
plt.title('Count by Class')
plt.show()
```

```python
sns.barplot(x='class', y='AREA', data=model_df)
plt.title('Area by Class')
plt.show()
```

```python
sns.barplot(x='class', y='AREA', data=model_df[model_df['class'] != 'none'])
plt.title('Area by Class (Less None)')
plt.show()
```

```python
sns.barplot(x='class', y='LENGTH', data=model_df)
plt.title('Length by Class')
plt.show()
```

```python
sns.barplot(x='class', y='NUM_PT', data=model_df)
plt.title('Number of Points by Class')
plt.show()
```

```python
sns.barplot(x='class', y='DIST_FROM_CENTER', data=model_df)
plt.title('Distance from Center (mm) by Class')
plt.show()
```

```python
sns.barplot(x='class', y='NUM_WAFER_PT', data=model_df)
plt.title('Number of Wafer Points by Class')
plt.show()
```

```python
sns.scatterplot(x='AREA', y='LENGTH', data=model_df)
plt.title('Scatter Plot of Length vs Area')
plt.show()
```

```python
sns.scatterplot(x='AREA', y='NUM_PT', data=model_df)
plt.title('Scatter Plot of Number of Points vs Area')
plt.show()
```

# 2. Analyze relationships by correlation matrix

```python
corr_df = model_df.drop(
    columns=['c_id', 'WAFER_ID']).corr()
corr_df.head()
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(corr_df)
plt.title('Correlation Matrix of Predictor Variables')
plt.show()
```
